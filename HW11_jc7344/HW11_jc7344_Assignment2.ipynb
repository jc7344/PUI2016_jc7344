2. cluster the NUMBER OF ESTABLISHMENTS time series with K-means in **a few** clusters (as discussed there is no real good, sound way to decide what a good number is here. try a few options, keeping in mind a few is more than a couple, but i recommand you stay within the single digit numbers)

```python
def elbow(data,K):
#data is your input as numpy form
#K is a list of number of clusters you would like to show.
    # Run the KMeans model and save all the results for each number of clusters
    KM = [KMeans(n_clusters=k).fit(data) for k in K]
    
    # Save the centroids for each model with a increasing k
    centroids = [k.cluster_centers_ for k in KM]

    # For each k, get the distance between the data with each center. 
    D_k = [cdist(data, cent, 'euclidean') for cent in centroids]
    
    # But we only need the distance to the nearest centroid since we only calculate dist(x,ci) for its own cluster.
    globals()['dist'] = [np.min(D,axis=1) for D in D_k]
    
    # Calculate the Average SSE.
    avgWithinSS = [sum(d)/data.shape[0] for d in dist]
    
    
    # elbow curve
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(K, avgWithinSS, 'b*-')
    plt.grid(True)
    plt.xlabel('Number of clusters')
    plt.ylabel('Average within-cluster sum of squares')
    plt.title('Elbow for KMeans clustering')
    plt.show()
    
    
    # Total with-in sum of square plot. Another way to show the result.
    wcss = [sum(d**2) for d in dist]
    tss = sum(pdist(data)**2)/data.shape[0]
    bss = tss-wcss
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(K, bss/tss*100, 'b*-')
    plt.grid(True)
    plt.xlabel('Number of clusters')
    plt.ylabel('Percentage of variance explained')
    plt.title('Elbow for KMeans clustering')
    plt.show()
```

```python
elbow(zc_shape.iloc[:, range(5,26)], range(1,30))
```

The figures above display the usage of the elbow method as a way to determine the optimal number of clusters for k-means. in this case this number is 2

```python
X=zc_shape.iloc[:, range(5,26)]
range_n_clusters = range(2,10)
for n_clusters in range_n_clusters:
    km = KMeans(n_clusters=n_clusters, random_state=324)
    cluster_labels = km.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters ={},".format(n_clusters)+" the average silhouette_score is :{}".format(silhouette_avg))
```

```python
mod_km_2 = KMeans(n_clusters=2, init='k-means++', n_init=10, max_iter=300, tol=0.0001, 
                verbose=0, random_state=None).fit(X)
```

```python
centers_2 = mod_km_2.cluster_centers_
```

```python
plt.figure(figsize=(8,8))
plt.scatter(zc_shape[mod_km_2.labels_ == 0].iloc[:,5], zc_shape[mod_km_2.labels_ == 0].iloc[:,24], \
           color='lightblue', alpha=0.3, label="Cluster 1")
plt.scatter(centers_2[0][0], centers_2[0][-1], marker='*', s=100, color='blue', label="Cluster 1 Center")

plt.scatter(zc_shape[mod_km_2.labels_ == 1].iloc[:,5], zc_shape[mod_km_2.labels_ == 1].iloc[:,24], \
           color='red', alpha=0.3, label="Cluster 2")
plt.scatter(centers_2[1][0], centers_2[1][-1], marker='*', s=100, color='green', label="Cluster 2 Center")

plt.xlabel("Establishments 1994")
plt.ylabel("Establishments 2014")
plt.title("KMeans Cluster: The Number Of Establishments 1994-2014\nClusters=2")
plt.legend(fontsize=10, loc='best')
```

```python
Z = linkage(zc_shape.iloc[:,range(5,26)], 'ward')
```

```python
plt.figure(figsize=(30,20))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,  
    leaf_font_size=8.,  
)
plt.show()
```

The hierarchical closter method builds the hierarchy from the individual elements by progressively merging clusters (wikkipedia)

```python
HC_label = fcluster(Z, 4, criterion='maxclust')
```

```python
zc_shape_2 = zc_shape.copy()
zc_shape_2['HC'] = HC_label
zc_shape_2['HC_'] = zc_shape_2.HC.replace([1,4],[4,1])
```

```python
f, ax = plt.subplots(figsize=(10,10))
zc_shape_2.plot(cmap='hot_r', column='HC_', ax=ax, categorical=True, k=4)
ax.set_title("Map 4: Hierarchical Clustering, k=4", fontsize=13)
ax.set_ylabel("Latitude")
ax.set_xlabel("Longitude")
```
